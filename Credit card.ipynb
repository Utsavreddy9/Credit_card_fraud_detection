```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv(r'D:\DS projects\Credit card\creditcard.csv')
```

```python
df['Class'].value_counts()
```

```python
# taking two variables for legit transactions and fraud transactions
legit = df[df.Class==0]
fraud = df[df.Class==1]
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# let's get a sample dataset from the original dataset 
legit_data = legit.sample(n=492)
```

```python
# now we will concatenate the fraud and legit_data dataframes
sample_data = pd.concat([legit_data,fraud],axis=0)
```

```python
sample_data
```

```python
#Splitting the data into features and targets

X = sample_data.drop(columns='Class',axis=1)
Y = sample_data['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
# splitting the data in train and test data
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
# MODEL TRAINING
```

```python
#using Logistic Regression
```

```python
model =  LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
#Testing the accuracy score of the model with train data
```

```python
X_train_prediction = model.predict(X_train)
accuracy_train_data = accuracy_score(X_train_prediction,Y_train)
```

```python
print(accuracy_train_data)
```

```python
#Testing the accuracy score of the model with test data
```

```python
X_test_prediction = model.predict(X_test)
accuracy_test_data = accuracy_score(X_test_prediction,Y_test)
```

```python
print(accuracy_test_data)
```
